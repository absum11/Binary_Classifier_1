#NUMERICAL ANALYSIS-
Probability density plots

```python
import matplotlib.pyplot as plt
def kdeplot(num_col):                                  # kdeplot gives a probability density graphical representation.
    plt.figure(figsize=(10,5))
    plt.title("KDE FOR {}".format(num_col))
    ax0=sns.kdeplot(df2[df2['Churn']==0][num_col].dropna(),color='red',label='Churn : No')
    ax1=sns.kdeplot(df2[df2['Churn']==1][num_col].dropna(),color='navy',label='Churn : Yes')
kdeplot('tenure')
kdeplot('MonthlyCharges')
kdeplot('TotalCharges')    
```

Some observations:
    1. recent customers more likely to churn
    2. customers with high monthly charges also likely to churn
